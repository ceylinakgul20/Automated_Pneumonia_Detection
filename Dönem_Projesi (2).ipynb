```python
!pip install -q kagglehub


import kagglehub
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from sklearn.metrics import classification_report, confusion_matrix

print("Veri seti indiriliyor... (Lütfen bekleyiniz)")
# KaggleHub ile şifresiz, otomatik indirme
path = kagglehub.dataset_download("paultimothymooney/chest-xray-pneumonia")

# Dosya yolunu dinamik bulma (İç içe klasör sorununa karşı)
dataset_path = path
if os.path.exists(os.path.join(path, "chest_xray")):
    dataset_path = os.path.join(path, "chest_xray")

print(f"Veri seti hazır: {dataset_path}")

# BÖLÜM 2: VERİ ÖN İŞLEME (Data Preprocessing)
# Kriter: Train/Val/Test ayrımı, Normalizasyon, Class Weight
# Amaç: Veriyi modele uygun hale getirmek
# - Train / Validation / Test ayrımı
# - Görüntü boyutlarının eşitlenmesi
# - Dengesiz veri (imbalanced data) problemi

BATCH_SIZE = 32  # Mini-batch boyutu
IMG_SIZE = (150, 150) # Görüntü boyutu standardizasyonu / Tüm görüntüler bu boyuta ölçeklenir


train_dir = os.path.join(dataset_path, 'train')
test_dir = os.path.join(dataset_path, 'test')

# 1. Eğitim ve Validasyon Ayrımı (%80 Train, %20 Validation)
# 1. Eğitim ve Validasyon Ayrımı
# validation_split ile eğitim verisinin %20’si validation olarak ayrılır
# seed sabitlenerek her çalıştırmada aynı bölünme elde edilir
# validation_split argümanı veriyi otomatik böler.

print("\n--- Veri Yükleniyor ---")
train_ds = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    validation_split=0.2,   # Kriter: Verinin ayrılması  # %80 train
    subset="training",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode='binary'# İkili sınıflandırma (Normal / Pneumonia)
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode='binary'
)

# 2. Test Verisi
test_ds = tf.keras.utils.image_dataset_from_directory(
    test_dir,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode='binary',
    shuffle=False # Test için karıştırma kapatıldı (Sıralı analiz için)
)

# 3. Performans Ayarı (Cache & Prefetch)
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)

# 4. Dengesiz Veri (Imbalanced Data) Yönetimi
# Normal ve Pnömoni sayılarını bulup ağırlık (Class Weight) hesaplıyoruz.
n_normal = len(os.listdir(os.path.join(train_dir, 'NORMAL')))
n_pneumonia = len(os.listdir(os.path.join(train_dir, 'PNEUMONIA')))
total = n_normal + n_pneumonia

weight_for_0 = (1 / n_normal) * (total / 2.0)
weight_for_1 = (1 / n_pneumonia) * (total / 2.0)
class_weights = {0: weight_for_0, 1: weight_for_1}

print(f"Sınıf Ağırlıkları -> Normal: {weight_for_0:.2f}, Pnömoni: {weight_for_1:.2f}")


# BÖLÜM 3: MODEL MİMARİSİ (CLASS YAPISI) ruveyda
# Amaç: Modelin sınıf yapısında oluşturulması, Normalizasyon, Augmentation


# Data Augmentation Bloğu (Overfitting olmasını önlemeye yarar)
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
])

class PneumoniaModel(tf.keras.Model):
    def __init__(self):
        super(PneumoniaModel, self).__init__()

        # Görüntüleri 0-255 aralığından 0-1 aralığına çeker (Normalization)
        self.rescaling = layers.Rescaling(1./255)

        # Kriter: Data Augmentation
        self.aug = data_augmentation

        # CNN Katmanları
        self.conv1 = layers.Conv2D(32, (3, 3), activation='relu', padding='same')
        self.bn1 = layers.BatchNormalization()
        self.pool1 = layers.MaxPooling2D((2, 2))

        self.conv2 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')
        self.bn2 = layers.BatchNormalization()
        self.pool2 = layers.MaxPooling2D((2, 2))

        self.conv3 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')
        self.bn3 = layers.BatchNormalization()
        self.pool3 = layers.MaxPooling2D((2, 2))

        # Sınıflandırma Katmanları (Fully Connected katmanlar)
        self.flatten = layers.Flatten()
        self.dense1 = layers.Dense(256, activation='relu')
        self.dropout = layers.Dropout(0.5) # Overfitting önleyici
        self.out = layers.Dense(1, activation='sigmoid') # Binary output

    def call(self, inputs, training=False):
        x = self.rescaling(inputs)

         # Augmentation sadece eğitim sırasında uygulanır
        if training:
            x = self.aug(x)

        x = self.conv1(x)
        x = self.bn1(x)
        x = self.pool1(x)

        x = self.conv2(x)
        x = self.bn2(x)
        x = self.pool2(x)

        x = self.conv3(x)
        x = self.bn3(x)
        x = self.pool3(x)

        x = self.flatten(x)
        x = self.dense1(x)

        # Dropout yalnızca training aşamasında aktiftir
        if training:
            x = self.dropout(x)
        return self.out(x)

# Modeli Başlatma / Model nesnesi oluşturulur
model = PneumoniaModel()
model.build(input_shape=(None, 150, 150, 3))
model.summary() # Kriter: Model özeti gösterimi

# Derleme (Compile)
# # Accuracy tek başına yeterli olmadığı için Recall metriği eklenmiştir.
# Tıbbi uygulamalarda FN (kaçırılan hasta) kritik olduğu için recall önemlidir.

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.Recall(name='recall')]
)


# BÖLÜM 4: EĞİTİM SÜRECİ (TRAINING)
# Kriter: Loss değerlerinin kaydedilmesi


print("\n--- Eğitim Başlıyor ---")
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=5, restore_best_weights=True
)

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=12,
    callbacks=[early_stopping],
    class_weight=class_weights, # Kriter: Dengesiz veri yönetimi
    verbose=1
)

# BÖLÜM 5: GÖRSELLEŞTİRME VE OVERFITTING İŞARETLEME
# Amaç:
# - Eğitim ve doğrulama performansını karşılaştırmak
# - Modelin ezber yapmaya (overfitting) başladığı noktayı tespit etmek

# Eğitim sürecinde her epoch için kaydedilen metrikler alınır
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(1, len(acc) + 1)

plt.figure(figsize=(15, 6))

# Accuracy Grafiği
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Train Accuracy')
plt.plot(epochs_range, val_acc, label='Val Accuracy')
plt.title('Accuracy Değişimi')
plt.legend(loc='lower right')

# Loss Grafiği (Overfitting İşaretlemeli)
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Train Loss')
plt.plot(epochs_range, val_loss, label='Val Loss', color='red', linestyle='--')

# En iyi epoch'u bul (Validation Loss'un en düşük olduğu an)
best_epoch_idx = np.argmin(val_loss)
best_epoch_val = val_loss[best_epoch_idx]
best_epoch_num = best_epoch_idx + 1

# En iyi noktayı işaretle
plt.annotate(f'En İyi Nokta\n(Epoch: {best_epoch_num})',
             xy=(best_epoch_num, best_epoch_val),
             xytext=(best_epoch_num, best_epoch_val + 0.2),
             arrowprops=dict(facecolor='green', shrink=0.05))

# Eğer eğitim devam ettiyse ve loss arttıysa orası Overfitting'dir
if best_epoch_num < len(loss):
    plt.annotate('Overfitting Başlangıcı',
                 xy=(best_epoch_num + 1, val_loss[best_epoch_idx + 1]),
                 xytext=(best_epoch_num + 2, val_loss[best_epoch_idx + 1] + 0.2),
                 arrowprops=dict(facecolor='red', shrink=0.05),
                 color='red')

plt.title('Loss Değişimi ve Overfitting Analizi')
plt.xlabel('Epochs')
plt.legend()
plt.show()


# BÖLÜM 6: TEST VE PERFORMANS RAPORU


print("\n--- Test Seti Sonuçları ---")
# Tüm test verisini ve etiketlerini tek seferde al (Hata analizi için gerekli)
all_images = []
all_labels = []
all_preds = []

print("Test verisi işleniyor...")
for images, labels in test_ds:
    preds = model.predict(images, verbose=0)
    all_images.append(images.numpy())
    all_labels.append(labels.numpy())
    all_preds.append(preds)

# Numpy array'e çevirme
X_test = np.concatenate(all_images)
y_test = np.concatenate(all_labels).flatten()
y_pred_prob = np.concatenate(all_preds).flatten()
y_pred_class = (y_pred_prob > 0.5).astype(int)

# Raporlama
print(classification_report(y_test, y_pred_class, target_names=['Normal', 'Pneumonia']))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_class)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Normal', 'Pneumonia'], yticklabels=['Normal', 'Pneumonia'])
plt.title('Confusion Matrix')
plt.show()


# BÖLÜM 7: HATA ANALİZİ (En Yüksek Güvenli Yanlışlar)
# Kriter: En yüksek güvenle yanlış bilinen 3 örneğin görselleştirilmesi


print("\n--- En Yüksek Güvenli 3 Hata (Top-3 High Confidence Errors) ---")

# Hata Skoru = |Tahmin Olasılığı - Gerçek Etiket|
# Örn: Gerçek 1, Tahmin 0.01 -> Fark 0.99 (Büyük Hata, Model "Normal" olduğundan çok emin)
error_diff = np.abs(y_test - y_pred_prob)

# Sadece yanlış tahminleri bul
wrong_indices = np.where(y_test != y_pred_class)[0]

if len(wrong_indices) > 0:
    # Yanlışlar içinden hata farkı en büyük olan 3 tanesinin indeksini bul
    # argsort küçükten büyüğe sıralar, [-3:] ile en sondaki en büyükleri alırız
    top_3_indices = wrong_indices[np.argsort(error_diff[wrong_indices])[-3:]]

    plt.figure(figsize=(15, 5))
    for i, idx in enumerate(top_3_indices):
        ax = plt.subplot(1, 3, i + 1)
        img = X_test[idx].astype("uint8")
        plt.imshow(img)

        true_lab = "Pneumonia" if y_test[idx] == 1 else "Normal"
        pred_lab = "Pneumonia" if y_pred_prob[idx] > 0.5 else "Normal"
        conf = y_pred_prob[idx] if y_pred_prob[idx] > 0.5 else 1 - y_pred_prob[idx]

        plt.title(f"Gerçek: {true_lab}\nTahmin: {pred_lab}\nGüven: %{conf*100:.1f}", color='red', fontweight='bold')
        plt.axis("off")
    plt.suptitle("Modelin En Çok Yanıldığı (En Güvendiği Yanlışlar) 3 Örnek")
    plt.show()


# BÖLÜM 8: TEKİL GÖRÜNTÜ TAHMİNİ (Single Image Prediction)
import numpy as np
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
from google.colab import files  # Google Colab dosya yükleme aracı

print("Lütfen analiz etmek istediğiniz röntgen görüntüsünü seçin...")
uploaded = files.upload() # Dosya seçme penceresini tetikler

for fn in uploaded.keys():
    path = fn

    # 1. Görüntüyü Yükleme ve Boyutlandırma
    # Modelin IMG_SIZE=(150, 150) ile eğitildiği için resmi bu boyuta zorluyoruz.
    img = image.load_img(path, target_size=(150, 150))

    # 2. Ön İşleme (Array Dönüşümü)
    x = image.img_to_array(img)

    # 3. Batch Boyutu Ekleme
    # Model (Batch_Size, Height, Width, Channels) bekler.
    # Elimizde tek resim olduğu için (150, 150, 3) -> (1, 150, 150, 3) yapıyoruz.
    x = np.expand_dims(x, axis=0)

    # 4. Tahmin (Prediction)
    # Modelin içinde Rescaling olduğu için manuel olarak 255'e bölmüyoruz.
    prediction = model.predict(x)
    score = prediction[0][0] # Çıktı örneğin [[0.85]] şeklindedir

    # 5. Sonucu Görselleştirme ve Yazdırma
    plt.figure(figsize=(4, 4))
    plt.imshow(img)
    plt.axis('off')

    print(f"\nDosya Adı: {fn}")
    print(f"Pnömoni Skoru: {score:.4f}")

    if score > 0.5:
        confidence = score * 100
        print(f"SONUÇ: ⚠️ PNEUMONIA (ZATÜRRE) - (%{confidence:.2f} Güven)")
        plt.title(f"PNEUMONIA (%{confidence:.1f})", color='red')
    else:
        confidence = (1 - score) * 100
        print(f"SONUÇ: ✅ NORMAL - (%{confidence:.2f} Güven)")
        plt.title(f"NORMAL (%{confidence:.1f})", color='green')

    plt.show()
```
